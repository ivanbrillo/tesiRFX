```python
kind = [
    {'mean': [0.2, 0.4], 'sigma': [0.1, 0.2], 'gain': [1.2, 1]},
    {'mean': [0.2, 0.4], 'sigma': [0.2, 0.2], 'gain': [1.5, 1]},
    {'mean': [0.2, 0.4], 'sigma': [0.1, 0.1], 'gain': [1.5, 1]}
]
```

```python
def gen_pt(size: int = 100, kind_index: int = 0):
    def gauss(x_points, mean, sigma, gain):
        return np.abs(
            np.exp(-np.power(x_points - mean, 2.) / (2 * np.power(sigma, 2.))) * gain)

    k = kind[kind_index]
    x = np.sort(np.random.rand(size))
    y = np.zeros_like(x)

    for m, s, g in np.stack([k['mean'], k['sigma'], k['gain']], axis=1):
        y += gauss(x, m, s, g)

    return np.stack([x, y], axis=1)
```

```python
from matplotlib import pyplot as plt
import numpy as np

plt.figure('data')
points = gen_pt(size=500, kind_index=2).transpose()
plt.plot(points[0], points[1])
```
